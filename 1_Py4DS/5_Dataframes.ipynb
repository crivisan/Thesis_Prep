```python
stats.InternetUsers < 40
```

```python
Filter = stats.InternetUsers < 2
```

```python
Filter
```

```python
stats[Filter]   #stats takes only the true values 
```

```python
#Awesome! LetÂ´s use this in practice now:
Filter2 = stats.BirthRate > 40
```

```python
stats[Filter2]
```

```python
stats[stats.BirthRate > 40]
```

```python
#even more interesting:
stats[stats.BirthRate > 40 and stats.InternetUsers < 2]  # error, The truth value of a Series is ambiguous.
```

```python
Filter and Filter2  #error, The truth value of a Series is ambiguous. #"Filters are Series"
```

```python
#Correcting this: 
Filter & Filter2
```

```python
stats[Filter & Filter2]
```

```python
stats[(stats.BirthRate > 40) & (stats.InternetUsers < 2)]   #must be the brackets otherwise Python does not understand
```

# 5.7. Using .at() and .iat() (advanced tutorial) 

```python
#Accessing individual elements
```

```python
stats[2,2] #error. Why??
#not so often used
```

```python
stats.head()
```

```python
# .at    #for labels. Important: even integers are treated as labels, "at that cell"
# .iat   #for integer location.
```

```python
stats.iat[3,4]   # .iat   = counts from left to right and top to bottom
```

```python
stats.iat[0,1]
```

```python
stats.at[2,"BirthRate"]    #Takes a label of the column
```

```python
# Why we need .at
```

```python
sub10 = stats[::10]
sub10
```

```python
sub10.iat[10,0]
```

```python
sub10.at[10,"CountryName"]     #it looks a country name (the label) with a number 10
```

# 5.9. Visualizing with Seaborn: Part 1

```python
vis3 = sns.lmplot(data=stats, x="InternetUsers", y="BirthRate", fit_reg = False)   
#linear model plot. # can be lmplot(x,y,data) or lmplot(data,x,y)
```

```python
vis3 = sns.lmplot(data=stats, x="InternetUsers", y="BirthRate", \
                  fit_reg = False, hue="IncomeGroup", size=8, aspect=1)    
#hue for the color 
#Size must be specified if needed, otherwise "lmplot" ignores "plt.rcParams['figure.figsize']=10,8" 
```
